```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#dataframe is a 2d datastructure with rows and columns
df = pd.read_csv('Top 50 Fast-Food Chains in USA.csv')
```

```python
df.head()
```

## Pairplot

```python
#https://seaborn.pydata.org/generated/seaborn.pairplot.html
g = sns.pairplot(df)
for ax in g.axes.flatten():
    ax.tick_params(rotation = 45)
plt.suptitle('Pair Plot for Top 50 Fast-Food Chains in USA, 2021', size = 20);
plt.show()
```

## Boxplot

```python
i=1
plt.figure(figsize=(10,10))
for col in df.select_dtypes("int"):
    plt.subplot(3,3,i) #i refers to which plot it is in the subplot
    sns.boxplot(df[col], orient="h") #orient can be h or v - default is v
    plt.title(col, size = 8)
    plt.tight_layout()
    i+=1
```

## Heatmap

```python
df1 = df
df2 = df1.drop("Total Change in Units from 2020",axis=1)
df2 = df2.drop("Average Sales per Unit (Thousands - U.S Dollars)",axis=1)
```

We clearly see a strong correlation between Systemwide Sales and Total Units. More interesting is the stronger correlation between systemwide sales and Franchised Stores than Company Stores, indicating that an increase in the number of Franchised Stores would have a higher impact on Systemwide Sales.

```python
sns.heatmap(df2.corr(), annot=True)
plt.show()
```
